```python
Medellin_population_covid = df[df['nombre_mun'] == 'Medellín']
```

```python
Medellin_population_covid_men = Medellin_population_covid[Medellin_population_covid['Sexo'] == 'M']
```

```python
len(Medellin_population_covid_men)
```

```python
Medellin_total_population = 2508452 # Dato cifra del DANE en 2018
```

```python
# Según la demografía de Medellín, el 46,6% de la población son hombres https://es.wikipedia.org/wiki/Demograf%C3%ADa_de_Medell%C3%ADn
Medellin_population_men = Medellin_total_population*0.466
Medellin_population_men
```

```python
# Teorema de Bayes
def Bayes(priorA, probBgivenA, probB):
    return round(((priorA * probBgivenA) / probB) * 100, 2)
```

```python
# Probabilidad de elegir a un hombre de la ciudad de Medellín y que este no contraiga COVID
priorA = Medellin_population_men/Medellin_total_population
probBgivenA = len(Medellin_population_covid_men)/Medellin_total_population

probB = (priorA * probBgivenA) + (((len(Medellin_population_covid)-len(Medellin_population_covid_men))/Medellin_total_population)*(1-priorA))

result = Bayes(priorA, probBgivenA, probB)
```
